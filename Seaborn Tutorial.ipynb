`Seaborn`, which specifically targets **statistical data visualization**, is complimentary to `Matplotlib`.

`Seaborn` works with different parameters, which allows users configure their plots instead of using the default looks of the `Matplotlib` plots.

Let's compare the following plots:
1. `Matplotlib` plot:

````python
# Import matplotlib and pandas
import matplotlib.pyplot as plt
import pandas as pd

# Initialize Figure and Axes object
fig, ax = plt.subplots()

# Load in data
tips = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Create violinplot
ax.violinplot(tips["total_bill"], vert=False)

# Show the plot
plt.show()
````

2\. `Seaborn` plot:

````python
# Import seaborn
import seaborn as sns

# Load the data
tips = sns.load_dataset("tips")

# Create violinplot
sns.violinplot(x = "total_bill", data=tips)

# Show the plot
plt.show()
````

`Matplotlib` doesn't work smoothly with `DataFrame`, especially when doing **exploratory analysis** with `Pandas`. With `Seaborn`, its plotting functions operate on DataFrames and arrays that contain a whole dataset.

Now let's load built-in dataset as example

````python
# Load iris data
iris = sns.load_dataset("iris")

# Construct iris plot
sns.swarmplot(x="species", y="petal_length", data=iris)

# Show plot
plt.show()
````
